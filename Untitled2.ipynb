### **Importar librerias**

```python
import csv
import keras
import numpy as np
```

### **Creacion del modelo**

```python
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='RMSprop', loss='mean_squared_error')
```

### **Obtener valores de CSV**

```python
with open('valores.csv', 'r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    Xfile=[]
    Yfile=[]
    
    for line in csv_reader:
        Xfile.append(line['x'])
        Yfile.append(line['y'])

    xs = np.array(Xfile, dtype=float)
    ys = np.array(Yfile, dtype=float)
    print(xs)
    print(ys)
```

### **Asignar epocas**

```python
model.fit(xs, ys, epochs=500)
```

## **Predecir**

```python
print(model.predict([51]))
print(model.predict([65]))
print(model.predict([80]))
```
